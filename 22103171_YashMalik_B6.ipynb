```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
city_x = np.array([100, 120, 85, 90, 110, 95])
city_y = np.array([80, 75, 60, 95, 85, 90])
city_z = np.array([150, 140, 135, 160, 155, 170])
```

```python
total_city_x = np.sum(city_x)
total_city_y = np.sum(city_y)
total_city_z = np.sum(city_z)

avg_city_x = np.mean(city_x)
avg_city_y = np.mean(city_y)
avg_city_z = np.mean(city_z)
```

```python
print(f"Total rainfall for city X: {total_city_x}")
print(f"Total rainfall for city Y: {total_city_y}")
print(f"Total rainfall for city Z: {total_city_z}")

print(f"Average rainfall for city X: {avg_city_x:.2f}")
print(f"Average rainfall for city Y: {avg_city_y:.2f}")
print(f"Average rainfall for city Z: {avg_city_z:.2f}")
```

```python
def average_across_cities(city_x, city_y, city_z):
    avg_monthly_rainfall = np.mean([city_x, city_y, city_z], axis=0)

    for month, avg_rainfall in enumerate(avg_monthly_rainfall, start=1):
        print(f"Month {month}: {avg_rainfall:.2f} mm")

average_across_cities(city_x, city_y, city_z)
```

```python
months = np.arange(1, 7)
plt.plot(months, city_x, label='city X',)
plt.plot(months, city_y, label='city Y',)
plt.plot(months, city_z, label='city Z',)
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Monthly Rainfall Trend for X, Y, Z')
plt.legend()
plt.show()
```

```python
range_city_x = np.max(city_x) - np.min(city_x)
range_city_y = np.max(city_y) - np.min(city_y)
range_city_z = np.max(city_z) - np.min(city_z)
```

```python
cities = ['City X', 'City Y', 'City Z']
ranges = [range_city_x, range_city_y, range_city_z]
print(ranges)
plt.bar(cities, ranges)
plt.xlabel('City')
plt.ylabel('Rainfall Range bar graph')
plt.title('Title: Rainfall range')
```
